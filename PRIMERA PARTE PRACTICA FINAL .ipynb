# **Sobre las CNN**

Las conocidas como Redes Neuronales Convoluciones (CNN, por sus siglas en inglés: Convolutional Neural Networks) consta de un tipo de arquitectura de redes de neuronas. CNNs se emplean en una amplia variedad de aplicaciones. Algunas de las aplicaciones más conocidas de esta Inteligencia Artificial (IA) pueden ser las siguientes:


-	Reconocimiento facial
-	Automóviles autónomos
-	Robots humanoides
-	Publicidad dirigida
-	Clasificación de imágenes
Ya en 1998, Yann LeCunn introdujo el concepto de Redes Neuronales Convolucionales (CNNs) en su artículo titulado “Gradient – Based Learning Applied to Document Recognition”.


Más tarde, en el año 2012, se popularizó con la ayuda de Alex Krizhevsky debido a su triunfo en la competición ImageNet, en la que utilizó las redes con el fin de reducir el error en la clasificación de un 26% hasta un 15%. Lo consiguió empleando técnicas básicas de visión por computador.

image.png

Este fue una disminución sustancial, por lo que dicho acontecimiento pasó a considerarse como un punto clave en la historia de la clasificación digital de imágenes. Se popularizó tiempo después debido a la insuficiente potencia de las GPUs de entonces.


El crecimiento exponencial a lo largo de los años de estas GPUs hizo posible el uso de redes más profundas (como las CNNs) 14 años más tarde, y hasta nuestros días. Actualmente, CNNs son empleadas en la gran mayoría de aplicaciones de ‘Deep Learning’ que se puedan encontrar.


Sin embargo, ¿qué es exactamente? La idea general es que es una red neuronal especial dedicada a procesar datos con una topología con estilo de malla (‘grid-like’). E posible imaginar una CNN como una red capaz de detectar patrones y sonsacar información (con sentido) de ellos. Es precisamente esta característica la que la hace tan útil para los análisis de imágenes.


Internamente, el proceso conlleva una ventana deslizante que reduce los diferentes aspectos de las imágenes en cuestión a menores dimensiones, a la vez que detecta dichos patrones. Este proceso se lleva a cabo empleando 4 tipos específicos de capas en su arquitectura:


-	**Capa Convolucional.** Tal y como explica su nombre, convoluciona la imagen utilizando filtros que detectan ejes, bordes, u otros componentes de la imagen (a grandes rasgos, sin entrar en detalles aún en las primeras capas, y profundizando en los detalles a medida que se va entrando en capas más profundas de la red). Estos filtros enseñan a la red durante el entrenamiento, de manera que dicha red aprenderá de forma autónoma valores de los filtros que detecten características y patrones relevantes para averiguar el ‘output’ correcto.
-	**Capa de Activación.** Su significado radica en la ausencia de aprendizaje en esta capa, aunque es esencial para las CNNs. Esta función es la transformación no lineal que se aplica a la señal de ‘input’. Varía con la aplicación, aunque algunas de las más utilizadas son la Sigmoid, ReLu y Leaky ReLU. La salida transformada (nuevo ‘output’) es enviada a la siguiente capa de neuronas como ‘input’ (capa de Pooling). Ayudan a a decidir si la neurona en cuestión mandaría (o no) su información a la siguiente capa.

image.png

image.png

-	**Capa de Pooling.** En una arquitectura CNN, estas capas se ubican entre las capas convolucionales. Reducen el número de parámetros y computación de la red, controlando el sobreajuste (‘overfitting’) al reducir progresivamente el tamaño espacial de dicha red. Normalmente, se utiliza “max-pooling”, que extraerá únicamente el máximo de los datos estudiados. A su vez, esto se consigue con filtros también, deslizando la ventana por el ‘input’ y extrayendo el parámetro máximo en cada iteración. El resto de parámetros se desechan.

image.png

Estas 3 capas mencionadas se pueden repetir múltiples veces en una misma red, a lo que llamaremos capas “conv”. De esta manera, se conseguirá una mayor profundidad en la red de manera progresiva (de ahí que se conozca como ‘Deep Learning’, o Aprendizaje Profundo). Un ejemplo de esquema podría ser el siguiente:

image.png

-	Por último, aparece la capa que se encuentra completamente conectada, y que aprende funciones no lineales provenientes de las salidas (‘ouputs’) de la última capa de Pooling. Aplana (‘flatten’) el volumen 3D que resulta de dicha cpa de Pooling a un vector 1D, que contendrá la misma cantidad total de parámetros. A continuación, se utiliza este vector en una red densamente conectada (completamente conectada), de una o más capas, tanto para la clasificación de imágenes como con otros fines. El resultado será un ‘output’ por imagen.

Nos encontramos con  un conjunto de datos **equilibrado**, donde cada especie está representada por una cantidad similar de imágenes. Nos puede ayudar a mejorar el rendimiento del modelo y evitar sesgos durante el entrenamiento.


Si existe una una cantidad desigual de imágenes por especie, podría resultar en un sesgo hacia las especies con más imágenes durante el entrenamiento. Esto puede hacer que el **modelo sea menos preciso** al clasificar especies con menos imágenes en comparación con las que tienen más.




Debemos mencionar que se trata de un conjunto de datos de tamaño reducido por lo que presenta ventajas y desventajas:

  - **Ventajas**:

    - **Menos recursos computacionales**: Con menos imágenes, el proceso de entrenamiento puede ser más rápido y requerir menos recursos computacionales, lo que es especialmente útil en sistemas con limitaciones de hardware.

    - **Menos riesgo de sobreajuste**: Con menos datos, existe un menor riesgo de sobreajuste, es decir, la red neuronal puede generalizar mejor para datos no vistos, siempre y cuando se maneje adecuadamente.

    - **Facilidad de recolección y etiquetado**: Recolectar y etiquetar un pequeño conjunto de datos puede ser menos costoso y menos laborioso que hacerlo para un conjunto de datos grande.

  - **Desventajas**:

    - **Menor capacidad de generalización**: Las CNN suelen funcionar mejor con conjuntos de datos más grandes y diversos. Una base de datos pequeña puede no proporcionar suficiente variedad para que la red aprenda características generales. Puede conllevar un sobreajuste del modelo.

    - **Mayor susceptibilidad al ruido**: Con menos datos, la red puede ser más sensible al ruido o a ejemplos atípicos en el conjunto de datos, lo que puede afectar negativamente su rendimiento.

    - **Limitaciones en la diversidad de clases o características**: Una base de datos pequeña puede no capturar la diversidad suficiente de clases o características presentes en el mundo real, lo que limita la capacidad de la red para reconocer y clasificar nuevas instancias.

```python
import os
import cv2

def load_images_from_folder(folder, label):
    images = []
    labels = []  # Prepare a list to store the labels of the images
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder, filename), cv2.IMREAD_UNCHANGED)
        if img is not None:
            img = cv2.resize(img, (64, 64))  # Resize images to uniform size
            images.append(img)
            labels.append(label)
    return images, labels

# Paths to your datasets
Lilly_path = '/content/flower_dataset/flower_images/Lilly'
Lotus_path = '/content/flower_dataset/flower_images/Lotus'
Orchid_path = '/content/flower_dataset/flower_images/Orchid'
Sunf_path = '/content/flower_dataset/flower_images/Sunflower'
Tulip_path = '/content/flower_dataset/flower_images/Tulip'
```

```python
import cv2
import os

# Load images and labels for each flower category
lilly_images, lilly_labels = load_images_from_folder(Lilly_path, 'Lilly')
lotus_images, lotus_labels = load_images_from_folder(Lotus_path, 'Lotus')
orchid_images, orchid_labels = load_images_from_folder(Orchid_path, 'Orchid')
sunf_images, sunf_labels = load_images_from_folder(Sunf_path, 'Sunflower')
tulip_images, tulip_labels = load_images_from_folder(Tulip_path, 'Tulip')
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Combine datasets
images = np.array(lilly_images + lotus_images + orchid_images + sunf_images + tulip_images)
labels = np.array(lilly_labels + lotus_labels + orchid_labels + sunf_labels + tulip_labels)

# Normalize pixel values
images = images / 255.0

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)

```

```python
# Reshape for CNN input
X_train = X_train.reshape(-1, 64, 64, 3)
X_test = X_test.reshape(-1, 64, 64, 3)
```

```python
from sklearn.preprocessing import LabelEncoder

# Convertir etiquetas de texto a enteros
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

```python
input_shape = X_train.shape[1:]
input_shape
```

## MODELO 1 CNN

Comenzamos con un modelo CNN

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Define the CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(5, activation='softmax')  # 5 classes for 5 flower types
])

# Compile the model
model.compile(optimizer='rmsprop',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Display the model summary
model.summary()
```

 Se muestra que la red tiene un total de 116,293 parámetros, todos ellos entrenables, lo que significa que se ajustarán durante el proceso de entrenamiento de la red para minimizar el error en la predicción o clasificación.

```python
from tensorflow.keras.callbacks import ModelCheckpoint
callbacks = [
    ModelCheckpoint(
        filepath="convnet_from_scratch.keras",
        save_best_only=True,
        monitor="val_loss")
]

# Train the model
history = model.fit(
    X_train,
    y_train_encoded,
    epochs=30,
    validation_data=(X_test, y_test_encoded),
    callbacks=callbacks
)
```

```python
# Evaluate the model on test data
test_model = keras.models.load_model("convnet_from_scratch.keras")
test_loss, test_accuracy = test_model.evaluate(X_test, y_test_encoded)

print(f'Test Loss: {test_loss}')
print(f'Test Accuracy: {test_accuracy}')

```

```python
# Plot training history for accuracy
plt.plot(history.history['accuracy'], label='train_accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training and Validation Accuracy')
plt.show()

# Plot training history for loss
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title('Training and Validation Loss')
plt.show()
```

- **Errores comunes**:

  - **Lotus y Lily**: Hay 14 instancias donde el modelo confundió Lotus con Lily, sugiriendo que estas dos clases pueden tener características visuales similares que confunden al modelo.

```python
# Cargar y preprocesar imágenes
images = [load_and_preprocess_image(path) for path in image_paths]

# Función para predecir y mostrar la imagen con la etiqueta y la probabilidad en dos filas
def predict_and_plot(model, images, labels, title):
    fig, axs = plt.subplots(2, len(images), figsize=(12, 8))  # Ajuste para dos filas
    fig.suptitle(title, fontsize=16)
    for i, image in enumerate(images):
        pred_probs = model.predict(image)
        pred_label = np.argmax(pred_probs, axis=1)
        pred_confidence = np.max(pred_probs) * 100

        # Mostrar la imagen
        axs[0, i].imshow(image[0])  # Ajusta si cambias la forma de cargar imágenes
        axs[0, i].axis('off')
        axs[0, i].set_title(f"Imagen {i+1}")

        # Mostrar la predicción
        axs[1, i].text(0.5, 0.5, f"{labels[pred_label[0]]}\n{pred_confidence:.2f}%",
                       horizontalalignment='center', verticalalignment='center', fontsize=12)
        axs[1, i].axis('off')
        axs[1, i].set_title("Predicción")

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# Etiquetas de las clases
labels = ['Lilly', 'Lotus', 'Orchid', 'Sunflower', 'Tulip']

# Cargar modelos preentrenados
model1 = tf.keras.models.load_model("convnet_from_scratch.keras")
model2 = tf.keras.models.load_model("convnet_from_scratch2.keras")
model3 = tf.keras.models.load_model("convnet_from_scratch_with_augmentation.keras")

# Realizar predicciones y mostrar resultados para cada modelo y cada imagen
predict_and_plot(model1, images, labels, "Resultados del Modelo 1")
predict_and_plot(model2, images, labels, "Resultados del Modelo 2")
predict_and_plot(model3, images, labels, "Resultados del Modelo 3")
```

La base de datos incluye imágenes de 1000 clases, que se dividen en 3 sets: entrenamiento (1.3M imágenes), validación (50K imágenes) y test (100K imágenes, con etiquetas de clase ‘hold-out’).


El desempeño de la clasificación es evaluado utilizando 2 medida:
-	Error top-1: error de clasificación multiclase. Por ejemplo, la proporción de imágenes clasificadas de manera incorrecta.
